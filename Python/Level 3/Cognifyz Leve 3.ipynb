```python
import pandas as pd
import matplotlib as plt
```

```python
Restaurant_Data = pd.read_csv("Cognifyz_Cleaned_Dataset.csv",encoding="utf-8",skiprows=0)
```

```python
# Creating a Copy of The Restaurant Data to avoid loss of Data During performing any irreversible operation

Restaurant_C_Data = Restaurant_Data.copy()
```

```python
import pandas as pd
import matplotlib.pyplot as pypolt
```

```python
#Level 3 Task 2

#Analyze if there is a correlation between the number of votes and the rating of a restaurant.
```

```python
import matplotlib.pyplot as pyplt
# Calculate correlation coefficient
correlation = Restaurant_C_Data["Aggregate rating"].corr(Restaurant_C_Data["Votes"])

# Print the correlation coefficient
print("Correlation coefficient between rating and votes:", correlation)

# Create a scatter plot
pyplt.figure(figsize=(8, 6))
pyplt.scatter(Restaurant_C_Data["Votes"], Restaurant_C_Data["Aggregate rating"])

# Add labels and title (using actual column names)
pyplt.xlabel('Number of Votes')
pyplt.ylabel('Aggregate Rating')
pyplt.title('Aggregate Rating vs. Number of Votes')

pyplt.grid(True)
pyplt.show()
```
