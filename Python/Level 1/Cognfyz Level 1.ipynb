```python
import pandas as pd
import matplotlib as plt
```

```python
Restaurant_Data = pd.read_csv("Cognifyz_Cleaned_Dataset.csv",encoding="utf-8",skiprows=0)
```

```python
# Creating a Copy of The Restaurant Data to avoid loss of Data During performing any irreversible operation

Restaurant_C_Data = Restaurant_Data.copy()
```

```python
Restaurant_C_Data.isnull()
```

```python
Restaurant_C_Data.isnull().sum()
```

```python
Restaurants_Count_For_Each_Range = Restaurant_C_Data["Price range"].value_counts()
```

```python
Restaurants_Count_For_Each_Range.index
```

```python
import matplotlib.pyplot as pyplt
# Assuming Restaurants_Count_For_Each_Range is a pandas Series or DataFrame
pyplt.bar(['1','2','3','4'], Restaurants_Count_For_Each_Range, color=['Red', 'green', 'cyan', 'magenta'])
pyplt.title('Bar Plot')
pyplt.xlabel('Price range')
pyplt.ylabel('Restaurants Count For Each Range')
pyplt.show()
```
